#Escape sequences

```python
print("Hello world \nHello panchu")           #Using \n to next line.
```

```python
print("Hello world \tHello panchu")          #Using \t to new tab.
```

```python
print("Hello \\ world")                   #Using \\ to backeslash.
```

```python
print("Hello \" world")                   #Using \" to double quote.
```

```python
print("Hello \' world")                    #Using \' to single quote.
```
